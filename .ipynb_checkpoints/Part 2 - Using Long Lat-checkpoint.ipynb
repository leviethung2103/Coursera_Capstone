```python
lat_list = []
long_list = []
for indx,data in enumerate(postal_code_df['Postcode']):
    lat = geo_df.loc[geo_df['PostalCode'] == data,'Latitude'].iloc[0]
    long = geo_df.loc[geo_df['PostalCode'] == data,'Longitude'].iloc[0]
    lat_list.append(lat)
    long_list.append(long)
```

```python
# Append the columns Latitude into postal_code_df
postal_code_df['Latitude'] = lat_list
postal_code_df['Longitude'] = long_list
```
